#### Solutions for question 3

For question 3, i used LSTM  RNN to train my prediction model.Basic neural nets such as RNN are faced with the vanishing gradient descend problems and LSTM solves this problem because it has a hidden state, and a memory cell with three gates namely forgotten, read and input gate.
Given a sequence of inputs, it will compute a sequence of hidden states.
I deceided to use str.comtains to filter out the payments_receipts that had the sequence MK in them. After that, I created a dataframe that contains only the payment receipts with the sequence MK and turned it into a csv file with the name data.csv.
Since this is a NLP problem, I used only the payment receipt row to train the model to predict the next character given an input of a sequence of charaters.
I therefore converted the payment_receipt row in the dataframe into a list and named it b and joined it with space. I then converted the list b into a textfile name listfile.txt

```python
#Load the data using pandas
import pandas as pd
df = pd.read_csv('train_revised.csv')
df
```

```python
print(df.columns.tolist())
```

```python
#Filter payment receipts to only contain those with the sequence MK

new = df['payment_receipt'].str.contains('MK') 
df[new]
```

```python
#Save the payment_receipts with sequence MK to a new csv file
df[new].to_csv('data.csv')
data = pd.read_csv('data.csv')
data.head()
```

```python
#Convert payment_receipt row into a lis then into a txt file
a = list(data['payment_receipt'])
# converting list into string and then joining it with space
b = ' '.join(str(e) for e in a)
  
# printing result
print(b)

#Convert list b to txt file
with open('listfile.txt', 'w') as filehandle:
    for listitem in b:
        filehandle.write('%s\n' % listitem)
```

```python
#Load the txt file into ascii then covert it to lower text
with open('listfile.txt', 'r') as f:
    data = f.read()
    data = data.lower()
    data = data.replace('\n', ' ')
    data = data.replace('\r', ' ')
    data = data.replace('\t', ' ')
    data = data.replace('  ', ' ')
#filename = "listfile.txt"
#raw_text = open(filename, 'r', encoding='utf-8').read()
#raw_text = raw_text.lower()

# create mapping of unique chars to integers
chars = sorted(list(set(data)))

char_to_int = dict((c, i) for i, c in enumerate(chars))
```

```python
#Summary of laoded data
print('Total Characters: ', len(data))
print('Unique Characters: ', len(chars))
print('Total Vocabulary: ', len(chars))
print('Unique Vocabulary: ', len(chars))

#n_chars = len(data)
#n_vocab = len(chars)
#print("Total Characters: ", n_chars)
#print( "Total Vocab: ", n_vocab)
```

```python
# prepare the dataset of input to output pairs encoded as integers
seq_length = 100
dataX = []
dataY = []
for i in range(0, len(data) - seq_length, 1):
	seq_in = data[i:i + seq_length]
	seq_out = data[i + seq_length]
	dataX.append([char_to_int[char] for char in seq_in])
	dataY.append(char_to_int[seq_out])
n_patterns = len(dataX)
print("Total Patterns: ", n_patterns)

```

```python
import numpy as np
from keras.utils import np_utils


# reshape X to be [samples, time steps, features]

X = np.reshape(dataX, (n_patterns, seq_length, 1))
# normalize
X = X / float(len(chars))


# one hot encode the output variable
y = np_utils.to_categorical(dataY)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras.layers import LSTM
from keras.callbacks import ModelCheckpoint

# define the LSTM model
model = Sequential()
model.add(LSTM(256, input_shape=(X.shape[1], X.shape[2])))
model.add(Dropout(0.2))
model.add(Dense(y.shape[1], activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics='accuracy')
```

```python
# define the checkpoint
filepath="weights-improvement-{epoch:02d}-{loss:.4f}.hdf5"
checkpoint = ModelCheckpoint(filepath, monitor='loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]
```

```python
#Train the model
model.fit(X, y, epochs=3, batch_size=128, callbacks=callbacks_list)
```

```python
#Summary of the model
model.summary()
```

```python
#Save the model
model.save('model.h5')



#model.save('model.h5')


```

```python
#Evaluate the model
import numpy as np
from keras.models import load_model
model = load_model('model.h5')
loss, acc = model.evaluate(X, y, verbose=0)
print("Model Accuracy: {:.2f}%".format(acc * 100))
print("Model Loss: {:.2f}".format(loss))
```
