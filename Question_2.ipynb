Since the question asks what is the probability that a passanger travelling from Kijauri will take a shuttle if they depart before 07.30, I took the count number of occurences when passangers will depart Kijauri before 07.30 and divided it by the count number of occurences where they car they is a shuttle.
To get the count number of occurneces when a passanger left Kijauri before 07.30, i used a for loop to loop through the entire travel_from row to get number of occurances for Kijauri and then used an if statement to filter to depatures that happended before 07.30.
I then looped through this to get the count of those that used a shuttle as their car_type. As for the dataset, i changed the travel_time format from HH:MM to HH.MM

```python
import pandas as pd
df = pd.read_csv('train_revised_q_two.csv')
df
```

```python
print(df.dtypes)
```

```python
travel_from = df['travel_from']
travel_time = df['travel_time']
df['travel_time'] < 7.30
car_type = df['car_type']

count = 0
#Probability that a passanger travelling from Kijauri will take a shuttle if they depart before 7:30

#Count the number who will depart Kijauri before 7:30
for i in range(len(travel_from)):
    if travel_from[i] == 'Kijauri' and travel_time[i] < 7.30:
        count += 1
print('The count for passangers who left Kijauri before 7.30 is: ',count)


#For those who will depart before 7:30, how many will take a shuttle?
count_shuttle = 0
for i in range(len(travel_from)):
    if travel_from[i] == 'Kijauri' and travel_time[i] < 7.30 and car_type[i] == 'shuttle':
        count_shuttle += 1
print('The count for those who left Kijauri before 7.30 and used a shuttle is: ',count_shuttle)







```

```python
#To get probability, divide the number of people who will depart before 7:30 by the total number of people who will depart before 7:30
probability = count_shuttle/count
print('The probability that a passanger who left Kijauri before 7.30 will take a shuttle is: ',probability)
```
