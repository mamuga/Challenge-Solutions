#### Solution.

For Question 1, I decided to add travel_route and travel_day rows. travel_route concatenates travel_from and travel_to rows into one and therefore shows the route travelled ie Migori-Nairobi routes.
travel_day shows the day in which a passanger travelled ie Sunday. For this, I changed the csv date format from DD-MM-YYYY to 
MM/DD/YYYY as the latter was easier to use in the csv file.
I then used the formula =TEXT(E2,"dddd") to add the day that corresponds to the date.
I saved the updated csv file as train_revised_added_travel_day_travel_route - train_revised.csv
I then used pyplot to plot an average_frequency bar graph and then ordered the graph in a descending order in order to rank the routes in a descending order.

```python
#Loading the data
import pandas as pd
df = pd.read_csv('train_revised_q_one.csv')
df
```

```python
#Filter the dataframe to only include Sunday in the row travel_day
df = df[df['travel_day'] == 'Sunday']
df
```

```python


# Calculate average route usage(an average frequency count of the routes used)
average_frequency = df['travel_route'].value_counts()/len(df)
average_frequency


```

```python

# Import required libraries
import matplotlib.pyplot as plt



#Plot the average frequency bar graph


plt.bar(average_frequency.index, average_frequency.values)
plt.title('Average Frequency of Travel Routes')
plt.xlabel('Travel Routes')
plt.ylabel('Frequency')
plt.show()
plt.grid()
plt.subplot()

#Order the top 7 routes by descending order of frequency in the graph
plt.bar(average_frequency.index[:7], average_frequency.values[:7])
plt.title('Top 7 Routes in decreasing order ')
plt.xlabel('Travel Routes')
plt.ylabel('Average Frequency')
plt.show()
plt.grid()










```
